# Exercise - K Means Clustering

In this exercise I will try to cluster universities into groups based off their features and see if Private and Public schools can be distinguished.

#### Step 1: Import libraries

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
```

#### Step 2: Make data
make_blobs can generate blob like groups of data and will be used to make some data

```python
from distutils.dep_util import newer_group
```

```python
data = make_blobs(n_samples=100, n_features=2, centers=4, cluster_std=1.8, random_state=101)
```

#### Step 3: Explore the data

```python
data
```

#### Step 4: Plot the created data

```python
plt.scatter(data[0][:,0],data[0][:,1])
plt.title('Fig. 1: Scatterplot')
```

According to the scatterplot in Fig. 1, there seems to be two blobs.

```python
plt.scatter(data[0][:,0],data[0][:,1], c=data[1])
plt.title('Fig. 2: Scatterplot')
```

#### Step 5: Create KMeans Clustering

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_ #predicted labels
```

#### Step 6: Compare the predicted values to the orignal values

```python
fig, (ax1,ax2) = plt.subplots(1,2, sharey=True, figsize=(10,6))

ax1.set_title('K Means')
ax1.scatter(data[0][:,0],data[0][:,1],c=kmeans.labels_,cmap='rainbow')

ax2.set_title('Original')
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```
